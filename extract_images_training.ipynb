```python
#### Extract jpgs for the training

import cv2
import sys
import os
import shutil

def clear_filepath(directory):
    for dirpath,_, filenames in os.walk(directory):
        for files in filenames:
            if not files.startswith("."):
                yield os.path.abspath(os.path.join(dirpath,files))

video_list_path = list(clear_filepath("videos/to_analyse/all"))
print(video_list_path)
start_frame = 1000
per_frame = 100

for video in video_list_path:
    out_filename = os.path.basename(video)
    print(out_filename)
    
    in_vid = cv2.VideoCapture(video)
    
    if not in_vid.isOpened():
        print("Cannot open input video file")
        sys.exit()
    
    ok, frame = in_vid.read()
    if not ok:
        print("Cannot read video file")
        sys.exit
    
    width, height = frame.shape[1], frame.shape[0]
    
    frame_nr = 0
    frame_counter = 0
    frame_string = str(frame_counter)
    
    while True:
        frame_nr += 1
        frame_string = str(frame_nr)
        
        if frame_nr > start_frame:
            if frame_counter == 0:
                img_name = out_filename.removesuffix(".mp4") + "_" + frame_string + ".jpg"
                dest_img = ("images/training/" + img_name)
                cv2.imwrite(dest_img, frame)
            if ((frame_counter % per_frame) == 0):
                img_name = out_filename.removesuffix(".mp4") + "_" + frame_string + ".jpg"
                dest_img = ("images/training/" + img_name)
                cv2.imwrite(dest_img, frame)
            frame_counter += 1
            
        ok, frame = in_vid.read()
        if not ok:
            break
    
    in_vid.release()

print("Done")
```
