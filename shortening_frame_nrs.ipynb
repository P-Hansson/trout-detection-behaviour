```python
# Create a video file with fewer frames if it has not already been done

import cv2
import sys
import os
import shutil
import time

def clear_filepath(directory):
    for dirpath,_, filenames in os.walk(directory):
        for files in filenames:
            if not files.startswith("."):
                yield os.path.abspath(os.path.join(dirpath,files))

filepath = "videos/raw"
video_list_path = list(clear_filepath(filepath))

filepath_done = "videos/to_analyse/all"
video_path2 = list(clear_filepath(filepath_done))
videos_done = [os.path.basename(n) for n in video_path2]

for video in video_list_path:
    out_filename = os.path.basename(video)
    if out_filename in videos_done:
        print(out_filename, "skip")
        continue
    
    print(out_filename)
    dest_in = ("videos/raw_done/" + out_filename)
    
    dest_out = ("videos/to_analyse/all/" + out_filename)
    
    in_vid = cv2.VideoCapture(video)
    
    if not in_vid.isOpened():
        print("Cannot open input video file")
        sys.exit()
    
    ok, frame = in_vid.read()
    if not ok:
        print("Cannot read video file")
        sys.exit

    width, height = frame.shape[1], frame.shape[0]
    
    fps = in_vid.get(cv2.CAP_PROP_FPS)
    
    out_vid = cv2.VideoWriter(out_filename, cv2.VideoWriter_fourcc(*"XVID"), fps, (width, height))
    
    frame_counter = 1
    
    while True:
        if ((frame_counter % 10) == 0):
            out_vid.write(frame)
        
        ok, frame = in_vid.read()
        if not ok:
            break
        
        frame_counter += 1
        
    out_vid.release()
    in_vid.release()
    
    shutil.move(out_filename, dest_out)
    shutil.move(video, dest_in)

print("Done")
```
